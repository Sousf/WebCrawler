```python
import pandas as pd
from bs4 import BeautifulSoup
import sys
import requests
from urllib.request import urlopen
from urllib.error import HTTPError
from urllib.error import URLError
from urllib.parse import urljoin
```

```python
# Requesting url and getting back html page.
seed = "https://www.goodfood.com.au/recipes/peanut-butter-and-chocolate-swirl-cheesecake-brownies-recipe-20120724-29tv8"
try:
    html = urlopen(seed)
except HTTPError as e:
    print("The server returned an HTTP error")
except URLError as e:
    print("The server could not be found!")
    
    
    
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Initalise dict and count of all websites that were visited
visited = {}
visited[seed] = True
pages_visited = 1

max_pages = 100

to_visit = []
links = soup.findAll('a')

# Find all embedded links within the first url.
for link in links:
    absolute_link = urljoin(seed, link['href'])
    if (absolute_link not in visited):
        to_visit.append(absolute_link)
        
        
```

```python
# Traverse the Beautiful Internet ;)
def ItsyBitsySpider(visited, to_visit, pages_visited, max_pages):
    while (to_visit and pages_visited < max_pages):
        # Taking out the link we currently will scrape.
        link = to_visit.pop(0)
        print(link)
        
        
        
        
        
        # Making request and soup to this current link.
        try:
            sub_page = requests.get(link)
            sub_soup = BeautifulSoup(sub_page.text, 'html.parser')
            
            
            # Do some stuff with the current link.
            
            
            
            # Make sure to not accidently DoS the web server!!!!!
            visited[link] = True
            new_links = sub_soup.findAll('a')

            for new_link in new_links:
                absolute_sublink = urljoin(link,new_link['href'])

                # Check that we don't have to revisit the same link again! To make sure no infinite loop
                if absolute_sublink not in visited and absolute_sublink not in to_visit:
                    to_visit.append(absolute_sublink)
            
            
            
            
        except Exception as e:
            print("Exception found on:", link, 'Exception:', e)
            visited[link] = True

        pages_visited += 1
```

```python
ItsyBitsySpider(visited, to_visit, pages_visited, max_pages)
```
